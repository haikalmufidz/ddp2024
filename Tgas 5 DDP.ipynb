```python
#no 2
print("program untuk menghitung luas bangun datar")
print("pilih menu angka 1-3 :\n1. persegi\n2. lingkaran\n3. segitiga")
PilihMenu = int(input("silahkan pilih menu:"))

match PilihMenu:
  case 1:
    print("menu untuk menghitung luas persegi")
    sisi = int(input("masukkan nilai yang ingin dihitung:"))
    hitung = sisi*sisi
    print(f"luas persegi adalah:{hitung}")
  case 2:
    print("menu untuk menghitung lingkaran")
    r = int(input("masukan nilai yang ingin dihitung:"))
    phi = 3.10
    hitung = phi*r*r
    print(f"luas lingkaran adalah:{hitung}")
  case 3:
    print("menu untuk menghitung segitiga")
    alas = int(input("masukan angka alas:"))
    tinggi = int(input("masukkan angka tinggi:"))
    hitung = 1/2*alas*tinggi
    print(f"luas segitiga adalah:{hitung}")
  case _:
    print("salah pilih menu")

```
